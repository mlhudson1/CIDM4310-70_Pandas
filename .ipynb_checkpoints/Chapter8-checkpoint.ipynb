## Understanding the basics of pandas visualization

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining a Dataframe
df = pd.DataFrame({
    'Year':range(2000,2020,1),
    'Sales':[4107,6492,1476,8508,7416,2747,1606,7947,9506,5441,7617,847,4389,3139,7546,3150,4426,4969,8457,5491]
})
df.Year=df.Year.astype('str')
df
```

```python
df.plot()
```

```python
df.plot();
```

```python
df.plot(x='Year');
```

```python
df.plot(x='Year', title= 'Yearly Sales');
```

```python
df.plot(x='Year', title='Yearly Sales', legend=False);
```

```python
df.plot(x='Year',title='Yearly Sales', legend = False, grid=True);
```

```python
df.plot(x='Year', title='Yearly Sales', legend = False, grid = True, color = 'tab:gray');
```

```python
df.plot(x="Year",title='Yearly Sales', grid=True, color='tab:gray', linestyle='dotted');
```

```python
df.plot(x='Year', title='Yearly Sales', gid=True, color='tab:gray', linestyle='dashed' );
```

```python
df.plot(kind='bar', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

```python
df.plot(kind='barh', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

```python
df.plot(kind='area', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

```python
df.plot(kind='pie', x = 'Year', y='Sales', title='Yearly Sales');
```

```python
df.plot(kind='scatter', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

## Exercise 8.01 - Building histograms for the Titanic dataset

## Exploring matplotlib

```python
# Defining a Dataframe
df = pd.DataFrame({
    'Year':range(2010,2020,1),
    'Sales':[4107,1606,7947,9506,5441,7617,8437,4389,3139,7546]
})
df.Year=df.Year.astype('str')

x=df['Year']
y=df['Sales']
plt.plot(x,y);
```

```python
x=df['Year']
y=df['Sales']
plt.title('Yearly Sales')
plt.xlabel('Years')
plt.ylabel('Sales in Units')
plt.plot(x,y);
```

```python
x=df['Year']
y=df['Sales']
plt.title('Yearly Sales')
plt.xlabel('Years')
plt.ylabel('Sales in Units')
plt.ylim(6000,10000)
plt.plot(x,y);
```

```python
x=df['Year']
y=df['Sales']
plt.title('Yearly Sales')
plt.xlabel('Years')
plt.ylabel('Sales in Units')
plt.plot(x,y, color = 'tab:gray');
```

```python
x=df['Year']
y=df['Sales']
plt.title('Yearly Sales')
plt.xlabel('Years')
plt.ylabel('Sales in Units')
plt.plot(x,y, color = 'tab:gray', linewidth = 5);
```

```python
x=df['Year']
y=df['Sales']
plt.title('Yearly Sales')
plt.xlabel('Years')
plt.ylabel('Sales in Units')
plt.plot(x,y, color = 'tab:gray', linewidth = 2, linestyle='--');
```

```python
x=df['Year']
y=df['Sales']
plt.title('Yearly Sales')
plt.xlabel('Years')
plt.ylabel('Sales in Units')
plt.plot(x,y, color = 'tab:gray', linewidth = 2, linestyle='--', marker='o');
```

## Visualizing data of different types

### Visualizing numerical data

```python
df=pd.DataFrame({
    'Height':[175,208,159,159,178,179,168,198,155,165,195,203,190,157,153,194,177,184,170,158]
})
df
```

```python
df.plot(kind='hist');
```

```python
# Defining a Dataframe
df = pd.DataFrame({
    'Weight':[67,75,119,69,106,111,120,80,108,100,79,53,75,89,120,67,70,77,65,71],
    'Height':[175,208,159,159,178,179,168,198,155,165,195,203,190,157,153,194,177,184,170,158]
})

df
```

### Visualizing Statistical Data

```python
df=pd.DataFrame({
    'Height':[175,208,159,130,178,179,168,100,155,165,195,250,190,157,153,194,177,184,170,210],
    'Gender':['F','M','F','F','M','M','F','M','F','F','M','M','F','M','M','M','F','M','M','F']
})
df
```

```python
df.boxplot(by="Gender", column="Height");
```

## Visualizing Multiple Data Plots

```python
df=pd.DataFrame({
    'Store':['A','A','A','A','A','A','A','A','A','A','B','B','B','B','B','B','B','B','B','B'],
    'Year':[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019],
    'Sales':[175,208,159,159,178,179,168,198,155,165,195,203,190,157,153,194,177,184,170,158]
})
df
```

```python
df.pivot(index='Year',columns='Store',values='Sales').plot();
```

```python
df.pivot(index='Year',columns='Store',values='Sales').plot(kind='bar');
```
