# Activity 14.01 - Analyzing air quality data

Consider that you're working as a data analyst for your city's municipality. The Department for the Environment needs your help in getting answeres to some questions related to emissions. The following are the questions the department wants answeres to:

    Which day of the week has the highest NO2(GT) emissions?
    
    At what time of the day are NMHC(GT) emissions highest?
    
    Which month has the lowest CO(GT) emissions?

1. Open a new Jupyter notebook.

```python
import pandas as pd
from dateutil.parser import parse
```

2. Download the data and then read the data using pandas.

```python
df=pd.read_csv('Chapter14-Datasets/AirQualityUCI.csv', sep=";")
df.head()
```

3. Drop unknown columns and rows with NA valus.

```python
df=df.drop(['Unnamed: 15','Unnamed: 16'],axis=1)
df.head()
```

```python
df.shape
```

```python
df=df.dropna()
df.shape
```

```python
df.info()
```

4. Extract the different attributes from the date column.

```python
df['Parse_date']=df['Date'].apply(lambda x: parse(x))
df['Weekday']=df['Parse_date'].apply(lambda x: x.weekday())
df['Day']=df['Parse_date'].apply(lambda x: x.strftime("%A"))
df['Month']=df['Parse_date'].apply(lambda x: x.strftime("%B"))
df.head()
```

5. Use the different methods you have learned in this chapter to answer all of the questions.

### Question 1: Which day of the week has the highest NO2(GT) emissions?

```python
Q1_1=pd.DataFrame(df.groupby(['Day'])['NO2(GT)'].agg('mean'))
Q1_1
```

### Question 2: At what time of the day are NMHC(GT) emissions highest?

```python
Q2_1=pd.DataFrame(df.groupby(['Time'])['NMHC(GT)'].agg('mean')) 
Q2_1
```

### Question 3: Which month has the highest CO(GT) emissions?

```python
def cleanFeat(x):
    return pd.to_numeric(".".join(x.split(',')))
```

```python
df['CO(GT)'] = df['CO(GT)'].apply(lambda x:cleanFeat(x))
df.head()
```

```python
Q3_1 = pd.DataFrame(df.groupby(['Month'])['CO(GT)'].agg('mean'))
Q3_1
```

6. Use suitable plotting methods to get relevent charts.

```python
DayPlot = Q1_1.plot.bar(y='NO2(GT)', rot=90,title='Emissions on week days')
DayPlot.set_xlabel("Week Days")
DayPlot.set_ylabel("Average Emissions")
```

```python
timeplot=Q2_1.plot.barh(y='NMHC(GT)', title = 'Emissions as hours of day')
timeplot.set_xlabel("Hours")
timeplot.set_ylabel("Average Emissions")
```

```python
monthPlot = Q3_1.plot.barh(y= 'CO(GT)',title = 'Monthly average emissions')

monthPlot.set_xlabel("Average Emissions")
monthPlot.set_ylabel("Months")
```
