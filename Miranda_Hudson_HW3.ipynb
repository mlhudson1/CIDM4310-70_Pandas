 * make the whole figure size as (15x5), i.e. each subplot will be 5x5

 * use the triangle marker `"^"`, marker inside (face) colors as `"blue"`, and marker edge colors =as `"black"`. Use the parameter `alpha=0.5` to make the markers semi-transparent

See the figure below as an example of the output

## <font color=Green> Task 1 Answer: </font>

```python
d=CarsData
figure1 = plt.figure(figsize = (15,5))
figure1.subplots(sharey='all')
var1plot = plt.subplot(1,3,1)
v1=var1plot.scatter(d["horsepower"],d["mpg"],marker="^",facecolors="blue",edgecolors="black", alpha=0.5)
var1plot.set_ylabel('MPG')
var1plot.set_xlabel('horsepower')
var1plot.yaxis.set_major_locator(ticker.MultipleLocator(5))
var2plot = plt.subplot(1, 3,2)
var2plot.set_xlabel('weight')
var2plot.tick_params('y', labelleft=False)
v2=var2plot.scatter(d["weight"],d["mpg"],marker="^",facecolors="blue",edgecolors="black", alpha=0.5)
var3plot = plt.subplot(1, 3,3)
var3plot.set_xlabel('cylinders')
var3plot.tick_params('y', labelleft=False)
v3=var3plot.scatter(d["cylinders"],d["mpg"],marker="^",facecolors="blue",edgecolors="black", alpha=0.5)
figure1.suptitle("Car MPG vs Horsepower, Weight, Cylinders")
plt.savefig('multi-scatters Of MPG vs HP_Wt_Cyl.pdf',bbox_inches='tight')
plt.show()
```

## <font color=Green> Task 2 Answer: </font>

```python
d=CarsData
figure1 = plt.figure(figsize = (15,8))
scales = np.linspace(3,8)
norm = plt.Normalize(scales.min(), scales.max())
min_max_scaler = preprocessing.MinMaxScaler()
wtNormal =min_max_scaler.fit_transform(d[["weight"]])*700;
heat=(1-0)*(d["cylinders"]-d["cylinders"].min()/(d["cylinders"].max()-d["cylinders"].min()))
cmap = plt.get_cmap("gist_yarg")
var1plot = plt.subplot(1,1,1)
v1=var1plot.scatter(d["horsepower"],d["mpg"],c=heat, cmap='rainbow', s=wtNormal, marker="o", edgecolors="black", alpha=0.5)
var1plot.set_ylabel('MPG')
var1plot.set_xlabel('horsepower')
var1plot.set_title("Car MPG vs HP, Weight(as marker size), and Cylinders(as color)")
sm =  ScalarMappable(norm=norm, cmap='rainbow')
sm.set_array([])
cbar = figure1.colorbar(sm)
cbar.set_label("Number of Cylinders")
plt.savefig('OneScatter Of MPG vs HP_Wt_Cyl.pdf',bbox_inches='tight')
plt.show()
```
