## Understanding the basics of pandas visualization

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining a Dataframe
df = pd.DataFrame({
    'Year':range(2000,2020,1),
    'Sales':[4107,6492,1476,8508,7416,2747,1606,7947,9506,5441,7617,847,4389,3139,7546,3150,4426,4969,8457,5491]
})
df.Year=df.Year.astype('str')
df
```

```python
df.plot()
```

```python
df.plot();
```

```python
df.plot(x='Year');
```

```python
df.plot(x='Year', title= 'Yearly Sales');
```

```python
df.plot(x='Year', title='Yearly Sales', legend=False);
```

```python
df.plot(x='Year',title='Yearly Sales', legend = False, grid=True);
```

```python
df.plot(x='Year', title='Yearly Sales', legend = False, grid = True, color = 'tab:gray');
```

```python
df.plot(x="Year",title='Yearly Sales', grid=True, color='tab:gray', linestyle='dotted');
```

```python
df.plot(x='Year', title='Yearly Sales', gid=True, color='tab:gray', linestyle='dashed' );
```

```python
df.plot(kind='bar', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

```python
df.plot(kind='barh', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

```python
df.plot(kind='area', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

```python
df.plot(kind='pie', x = 'Year', y='Sales', title='Yearly Sales');
```

```python
df.plot(kind='scatter', x = 'Year', y='Sales', title='Yearly Sales', color='tab:gray');
```

## Exercise 8.01 - Building histograms for the Titanic dataset

```python
file_path="Chapter8-Datasets/titanic.csv"
titanic=pd.read_csv(file_path)
titanic.head()
```

```python
titanic=titanic.dropna()
titanic
```

```python
titanic.age.plot(kind='hist',title='Histogram plot for ages of the passengers onboard Titanic');
```

```python
titanic.loc[titanic['survived']==1].age.plot(kind='hist',title='Histogram plot for ages of the passengers who survived Titanic');
```
